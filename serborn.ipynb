# Barplot:

A bar plot epresents an aggregate or statistical estimate for a numeric variable with the height of each rectangle and indicates the uncertainty around that estimate using an error bar. Bar plots include 0 in the axis range, and they are a good choice when 0 is a meaningful value for the variable to take

# Boxplot:

A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be â€œoutliersâ€ using a method that is a function of the inter-quartile range.


```python
sns.set(style='whitegrid')
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.boxplot(data=fmri, y="signal", x="timepoint")
```

# Violin plot:

A violin plot plays a similar role as a box-and-whisker plot. It shows the distribution of data points after grouping by one (or more) variables. Unlike a box plot, each violin is drawn using a kernel density estimate of the underlying distribution.

```python
import seaborn as sns
sns.set(style='whitegrid')
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.violinplot(data=tips, y="tip",x="time")
```

# Heatmap:

A heatmap in seaborn is a great way to visualize matrix-like data, often used for showing correlation matrices, distances, or any 2D data. It uses color coding to represent values in a 2D space, with darker or lighter colors indicating higher or lower values, depending on the color palette.

# Key Parameters for sns.heatmap()

# data:
Type: array-like or DataFrame-
Description: The data you want to plot. It should be a 2D dataset such as a pandas DataFrame or a numpy array. The values in the dataset are what will be plotted as colored cells.
Example: data=iris.corr()

# annot:
Type: bool or array-like

Description: If True, each cell will be annotated with its numerical value. You can also pass a 2D array-like object where each value corresponds to a specific annotation.

Default: False
Example: annot=True (will display the values in each cell)

# cmap:
Type: str or matplotlib colormap
Description: The color map to use for the heatmap. This controls the gradient of colors used to represent the values in the heatmap.
Default: None (uses the default color map)
Common Color Maps: 'viridis', 'coolwarm', 'Blues', 'YlGnBu', 'magma', 'cividis'
Example: cmap='coolwarm'

# vmin / vmax:
Type: float, int
Description: These parameters set the scaling range of the color map. vmin is the minimum value that corresponds to the lowest color in the color map, and vmax is the maximum value that corresponds to the highest color.
Default: None (automatically scales based on data range)
Example: vmin=0, vmax=1 (scales the color map to range from 0 to 1)

# center:
Type: float
Description: This parameter centers the colormap at a specific value. It is useful for diverging color maps (e.g., coolwarm) where you want to highlight values around a central point.
Example: center=0 (useful for correlation matrices where 0 represents no correlation)

# linewidths:
Type: float
Description: The width of the lines that will divide each cell in the heatmap.
Default: 0
Example: linewidths=0.5

# linecolor:
Type: str
Description: The color of the lines that separate the cells in the heatmap.
Default: 'white'
Example: linecolor='gray'

# mask:
Type: array-like or bool
Description: Mask a part of the heatmap by passing a bool array (where True will mask the cell) or a specific condition (like NaN values).
Example: mask=np.triu(correlation_matrix) (to mask the upper triangle of a correlation matrix)

# cbar:
Type: bool
Description: Whether to display the color bar (legend) alongside the heatmap.
Default: True
Example: cbar=False (disables the color bar)

# cbar_kws:
Type: dict
Description: Dictionary of additional keyword arguments to pass to the colorbar (e.g., shrink, ticks, etc.).
Example: cbar_kws={"shrink": 0.8, "ticks": [0, 0.5, 1]}

# xticklabels / yticklabels:
Type: bool, int, list of str, or list of int
Description: These control the appearance of the x and y-axis labels.
If True, labels are displayed.
If an integer, it sets the number of ticks to display.
If a list, it will use the values as custom labels.
Example: xticklabels=True (display labels automatically)
Example: yticklabels=5 (display every 5th label)

# fmt:
Type: str
Description: The format string used to display annotations in each cell.
Default: None (use the default string formatting)
Example: fmt='.2f' (formats annotations to show 2 decimal places)

# square:
Type: bool
Description: Whether to make the heatmap square (i.e., equal width and height for each cell).
Default: False
Example: square=True

# robust:
Type: bool
Description: If True, robust scaling is used for the color mapping. This can be useful when you have outliers in your data.
Default: False
Example: robust=True

# seaborn.catplot:

This function provides access to several axes-level functions that show the relationship between a numerical and one or more categorical variables using one of several visual representations. The kind parameter selects the underlying axes-level function to use.

```python
df = sns.load_dataset("titanic")
sns.catplot(data=df, x="age", y="class")
```

```python
#Use kind to select a different representation:

sns.catplot(data=df, x="age", y="class", kind="box")
```

```python
#One advantage is that the legend will be automatically placed outside the plot:

sns.catplot(data=df, x="age", y="class", hue="sex", kind="boxen")
```

```python
#Additional keyword arguments get passed through to the underlying seaborn function:

sns.catplot(
    data=df, x="age", y="class", hue="sex",
    kind="violin", bw_adjust=.5, cut=0, split=True)
```

```python
#Assigning a variable to col or row will automatically create subplots. Control figure size with the height and aspect parameters:

sns.catplot(
    data=df, x="class", y="survived", col="sex",
    kind="bar", height=4, aspect=.6)
```

```python
#For single-subplot figures, it is easy to layer different representations:


sns.catplot(data=df, x="age", y="class", kind="violin", color=".9", inner=None)
sns.swarmplot(data=df, x="age", y="class", size=3)
```

```python
#Use methods on the returned FacetGrid to tweak the presentation:


g = sns.catplot(
    data=df, x="who", y="survived", col="class",
    kind="bar", height=4, aspect=.6)
```

# seaborn.pointplot:

A point plot represents an estimate of central tendency for a numeric variable by the position of the dot and provides some indication of the uncertainty around that estimate using error bars.

Point plots can be more useful than bar plots for focusing comparisons between different levels of one or more categorical variables. They are particularly adept at showing interactions: how the relationship between levels of one categorical variable changes across levels of a second categorical variable. The lines that join each point from the same hue level allow interactions to be judged by differences in slope, which is easier for the eyes than comparing the heights of several groups of points or bars.



```python
df = sns.load_dataset("penguins")
df.head()
```

```python
import matplotlib.pyplot as plt

penguins = sns.load_dataset('penguins')
sns.pointplot(x='species', y='bill_length_mm', data=penguins)

plt.title('Average Bill Length by Penguin Species')
plt.xlabel('Species')
plt.ylabel('Average Bill Length (mm)')
plt.show()
```

```python
sns.pointplot(data=penguins, x="island", y="body_mass_g", errorbar="sd")
```
