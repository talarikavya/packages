# Matplotlib Plotting
*Plotting x and y points:-


The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.

# Matplotlib Bar plot:-

The bar plots are vertical/horizontal rectangular graphs that show data comparison where you can gauge the changes over a period represented in another axis (mostly the X-axis). Each bar can store the value of one or multiple data divided in a ratio. The longer a bar becomes, the greater the value it holds. In Matplotlib, we use the bar() or barh() function to represent.



```python
import matplotlib.pyplot as plt
```

```python
Elements = ['a','b','c','d']
values = (6,5,8,4)
color = 'green'

plt.bar(Elements,values,color = color)
plt.title("Bar plot example")
plt.ylabel("values")
plt.show()
```

# Matplotlib Pie plot:-

A pie plot is a circular graph where the data get represented within that components/segments or slices of pie. Data analysts use them while representing the percentage or proportional data in which each pie slice represents an item or data classification. In Matplotlib, the pie() function represents it.



```python
labels = ['A','B','C']
size = [88,77,66]

plt.pie(size,labels=labels)
plt.title('pie chart example')
plt.show()
```

```python
labels = ['Apple', 'Banana', 'Cherry', 'Orange']
sizes = [15, 30, 45, 10]  # These are the sizes for each section


plt.pie(sizes, labels=labels, autopct='%1.1f%%')      #autopct is used to percentage on each slice, formatted to one decimal place.
plt.title('Fruit Distribution')
plt.show()
```

```python
labels = ['iphone','realme','redmi','oneplus','noike']
sizes = [400,392,399,321,100]

plt.pie(sizes, labels=labels, autopct='%1.5f%%')
plt.title("moilbe price")
plt.show()
```

# Matplotlib Histogram plot:-

We can use a histogram plot when the data remains distributed, whereas we can use a bar graph to compare two entities. Both histogram and bar plot look alike but are used in different scenarios. In Matplotlib, the hist() function represents this.

# Matplotlib step plot:-

A step plot is a type of plot in matplotlib where the data points are connected with horizontal and vertical lines, creating a "staircase" effect. Its useful when you want to represent discrete data or visualize the change in values step by step.

```python
import numpy as np
```

```python
x = np.arange(1,5)
y = [1,2,3,4]

plt.step(x,y,label='stepplot',color='red',linewidth=3)
plt.title("step plot example")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
x = np.arange(1,9)
y = [4,5,2,6,7,1,3,5]

plt.step(x,y,label='stepplot',color='pink',linewidth=6)
plt.title("step plot example")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

# Matplotlib Bubble plot:-

A bubble plot is a variation of a scatter plot where each data point is represented by a bubble. The size of each bubble represents a third dimension of data, in addition to the x and y coordinates. This can be useful for visualizing three variables at once.

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50)*100
colors = np.random.rand(50)

plt.scatter(x,y,s=sizes,c=colors,alpha=0.5,cmap='plasma')
plt.colorbar(label='colors')
plt.title("bubble example")
plt.show()
```

```python
x = np.random.rand(1000)
y = np.random.rand(1000)
sizes = np.random.rand(1000)*100
colors = np.random.rand(1000)

plt.scatter(x,y,s=sizes,c=colors,alpha=0.9,cmap='magma')
plt.colorbar(label='colors')
plt.title("bubble example")
plt.show()
```

# Matplotlib Sub-plot :



# Matplotlib Creating Subplots with plt.subplots():

This method will remove everything from the axes: data, titles, labels, legends, etc., but will keep the axes object intact (so it can be reused).

```python
x = [39,38,78,56,53,89,67]
y = [78,66,45,88,25,87,45]       #Creating a 1x1 Grid of Subplots

plt.subplot(1, 1, 1)             #first row and first column
plt.plot(x, y, label='y = x^2')
plt.title('Single Plot: y = x^2')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show
```

```python
x=[45,23,55,44,77]
y=[44,88,66,34,55]

fig, ax = plt.subplots(1, 1, figsize=(6, 4))
ax.plot(x, y, label='y = x^2')                    #here ax is an object and fig,ax is figure object
ax.set_title('Single Plot: y = x^2')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.legend()
plt.show()
```

```python
import numpy as np

fig,a =plt.subplots(2,2)
x = np.arange(2,6)
a[0][0].plot(x,x*x)
a[0][0].set_title('square')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('square root')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('exp')
a[1][1].plot(x,np.log10(x))
a[1][1].set_title('log')
plt.show()
```

```python
#Creating a Simple Line Plot

import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

fig, ax = plt.subplots()    ## Create the figure and axis
plt.plot(x, y, label='y = x^2', color='blue')  ## Plot the data
plt.title('Simple Line Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()
```

```python
#create a 2x2 grid of subplot

x = [0, 1, 2, 3, 4, 5]
y1 = [0, 1, 4, 9, 16, 25]
y2 = [0, -1, -4, -9, -16, -25]
y3 = [0, 2, 4, 6, 8, 10]
y4 = [5, 6, 7, 8, 9, 10]

plt.subplot(2, 2, 1) #first row and colunm
plt.plot(x, y1, label='y = x^2')
plt.title('Plot 1: y = x^2')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()
```

```python
plt.subplot(2, 2, 2)      #first row second colunm
plt.plot(x, y2, label='y = -x^2', color='red')
plt.title('Plot 2: y = -x^2')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()
```

```python
plt.subplot(2, 2, 3)   #second row first colunm
plt.plot(x, y3, label='y = 2x', color='green')
plt.title('Plot 3: y = 2x')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()
```
